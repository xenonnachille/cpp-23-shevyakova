```python
import pandas as pd
df = pd.read_csv("binsearch.csv")
df
```

```python
from matplotlib import pyplot as plt
import numpy as np
N = np.array(df["N"])
search = np.array(df["time"])
fig, ax = plt.subplots(figsize=(6, 7))
ax.scatter(N, search, s=1)
ax.set(xlabel='$N$',
       ylabel='search: t, ns')
```
