```python
import pandas as pd
df = pd.read_csv("linsearch.csv")
df
```

```python
from matplotlib import pyplot as plt
import numpy as np
N = np.array(df["N"])
search = np.array(df["time"])
fig, ax = plt.subplots(figsize=(6, 7))
ax.scatter(N, search, s=1)
ax.set(xlabel='$N$',
       ylabel='search: t, ns')
```

```python
df2 = pd.read_csv("linsearch_extratime.csv")
```

```python
N2 = np.array(df2["N"])
search2 = np.array(df2["time"])
fig, ax = plt.subplots(figsize=(6, 7))
ax.scatter(N2, search2, s=1)
ax.set(xlabel='$N$',
       ylabel='search: t, ns')
```

```python
fig, ax = plt.subplots(figsize=(6, 7))
ax.scatter(N, search, s=0.1, color='pink')
ax.scatter(N2, search2, s=0.1, color='black')
ax.set(xlabel='$N$',
       ylabel='search: t, ns')
```
