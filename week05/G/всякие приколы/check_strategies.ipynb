```python
import pandas as pd
df = pd.read_csv("strategyA_.csv")
df
```

```python
from matplotlib import pyplot as plt
import numpy as np
N = np.array(df["N"])
tr = np.array(df["time_r"])
tn = np.array(df["time_n"])
fig, ax = plt.subplots(figsize=(6, 7))
ax.scatter(N, tr, s=1)
ax.set(xlabel='$N$',
       ylabel='search: t, ns')
```
