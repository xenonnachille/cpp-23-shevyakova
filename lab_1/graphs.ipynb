```python
import pandas as pd
df = pd.read_csv("search_times.csv")
df
```

```python
from matplotlib import pyplot as plt
import numpy as np
N = np.array(df["N"])
search = np.array(df["time_search"])
binsearch = np.array(df["time_binsearch"])
fig, ax = plt.subplots(figsize=(6, 7))
ax.scatter(N, search, s=3)
ax.set(xlabel='$N$',
       ylim=(-10, 1200), ylabel='игииду search: t, mks')
```

```python
import math
fig, ax = plt.subplots(figsize=(6, 7))

ax.scatter(N, binsearch, s=10)
ax.set(xlim=(-10000, 550000), xlabel='$N$',
       ylim=(20, 160), ylabel='binary search: t, mks')
x = np.arange(100, 700000, 1000)
y = [7.2*math.log(a) for a in x]
ax.plot(x, y, color='black', linestyle='--')
plt.show()
```
